```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Год': [1996, 2000, 2004, 2008, 2012, 2016, 2020],
    'Город': ['Атланта', 'Сидней', 'Афины', 'Пекин', 'Лондон', 'Рио-де-Жанейро', 'Токио'],
    'Спортсмены': [96, 130, 119, 132, 115, 104, 93],
    'Золото': [3, 3, 1, 2, 3, 2, 0],
    'Серебро': [4, 4, 4, 3, 2, 5, 0],
    'Бронза': [4, 0, 3, 4, 6, 10, 8],
    'Всего медалей': [11, 7, 8, 9, 11, 17, 8],
    'Место': [24, 22, 40, 29, 24, 30, 83]
}

df = pd.DataFrame(data)

df.set_index('Год')[['Золото', 'Серебро', 'Бронза']].plot(kind='bar', stacked=True)
plt.title('Количество медалей Казахстана по годам')
plt.xlabel('Год')
plt.ylabel('Количество медалей')
plt.show()
```

```python

plt.plot(df['Год'], df['Золото'], label='Золото', marker='o')
plt.plot(df['Год'], df['Серебро'], label='Серебро', marker='o')
plt.plot(df['Год'], df['Бронза'], label='Бронза', marker='o')
plt.title('Количество медалей по типам')
plt.xlabel('Год')
plt.ylabel('Количество медалей')
plt.legend()
plt.show()
```

```python

plt.plot(df['Год'], df['Спортсмены'], label='Количество спортсменов', marker='o')
plt.title('Количество спортсменов по годам')
plt.xlabel('Год')
plt.ylabel('Количество спортсменов')
plt.show()
```

```python

plt.plot(df['Год'], df['Место'], label='Место в мировом рейтинге', marker='o')
plt.title('Изменение места в мировом рейтинге')
plt.xlabel('Год')
plt.ylabel('Место')
plt.gca().invert_yaxis()  # Инвертируем ось Y, так как 1 место лучше 100-го
plt.show()
```

```python

df.plot(x='Год', y=['Всего медалей', 'Спортсмены'], kind='bar')
plt.title('Общее количество медалей и количество спортсменов')
plt.xlabel('Год')
plt.ylabel('Количество')
plt.show()
```

```python
medals_df = pd.melt(df, id_vars=['Год', 'Город'], value_vars=['Золото', 'Серебро', 'Бронза'], var_name='Тип медали', value_name='Количество')
medals_df.boxplot(by='Тип медали', column='Количество', grid=False)
plt.title('Распределение медалей по типам')
plt.suptitle('')
plt.xlabel('Тип медали')
plt.ylabel('Количество')
plt.show()
```

```python
total_medals = df[['Золото', 'Серебро', 'Бронза']].sum()
total_medals.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Доля каждого типа медалей от общего числа')
plt.ylabel('')
plt.show()
```
